```python
import pandas as pd

# Corrected file path - you need to verify this path exists on your system
# Option 1: Use the correct path to your file
file_path = r"C:\Users\srsch\Downloads\Medicare Part D Prescribers - by Provider and Drug\Medicare Part D Prescribers - by Provider and Drug\2023\MUP_DPR_RY25_P04_V10_DY23_NPIBN.csv"


# Read the CSV file once you've confirmed the path is correct
df = pd.read_csv(file_path, low_memory=False)
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
for col in df.columns:
    print(f"\n{col} value counts:")
    print(df[col].value_counts(dropna=False).head(10))
```

```python
### Semantic Cleaning Notes
#-`GE65_Sprsn_Flag`: Contains suppression indicators (`*`, `#`) and missing values. These will be flagged and excluded from beneficiary-level analysis.
#- `Prscrbr_Type_Src`: Multiple sources (`Claim-Specialty`, `NPPES`) suggest potential inconsistency. Will prioritize `Claim-Specialty` for modeling.
#- `Gnrc_Name` vs `Brnd_Name`: Will normalize and map to a unified drug taxonomy for clustering.
```

```python
# Flag suppression markers
suppression_flags = ['*', '#', 'NaN', 'nan', '', None]
for col in df.columns:
    df[col + '_flagged'] = df[col].astype(str).apply(lambda x: 'suppressed' if x in suppression_flags else 'valid')
```

```python
#Unified (try)brand and generic drug names into a single cleaned column for clustering &volume analysis
df['Drug_Name_Cleaned'] = df['Gnrc_Name'].combine_first(df['Brnd_Name'])
df['Drug_Name_Cleaned'] = df['Drug_Name_Cleaned'].str.lower().str.strip()
```

```python
#Prescriber Clusterin prep..for modeling and other fun with features
features = ['Tot_Clms', 'Tot_Drug_Cst', 'Tot_30day_Fills']
df_cluster = df[features].dropna()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=42)
df_cluster['cluster'] = kmeans.fit_predict(df_cluster)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select continuous columns and drop rows with NaNs
continuous_cols = ['Tot_Clms', 'Tot_30day_Fills', 'Tot_Day_Suply', 'Tot_Drug_Cst', 'Tot_Benes',
                   'GE65_Tot_Clms', 'GE65_Tot_30day_Fills', 'GE65_Tot_Drug_Cst', 'GE65_Tot_Day_Suply', 'GE65_Tot_Benes']

df_continuous = df[continuous_cols].dropna()

# Compute correlation matrix
corr_matrix = df_continuous.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Continuous Medicare Features")
plt.show()
```

```python
df['fills_delta'] = df['Tot_30day_Fills'] - df['GE65_Tot_30day_Fills']
df['fills_delta'].describe()
```

```python
df['fills_delta_flag'] = pd.cut(df['fills_delta'], bins=[-1, 0, 20, 100, 1000, df['fills_delta'].max()],
                                labels=['No delta', 'Small delta', 'Moderate', 'Large', 'Extreme'])
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['fills_delta'], bins=50, kde=True, color='teal')
plt.title("Distribution of Fills Delta (Total vs GE65)")
plt.xlabel("Delta in 30-Day Fills")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x='Prscrbr_Type', y='Tot_Drug_Cst', data=df)
plt.xticks(rotation=45)
plt.title("Drug Cost Distribution by Prescriber Type")
plt.ylabel("Total Drug Cost")
plt.show()
```
